```python
bact = df['CianoBacterias']
taxa = []

for i in range(len(bact)):
    ncel = bact[i]
    ##print(ncel)
    if ncel <= 20000:
        taxa.append(1.0)
    elif ncel > 20000 and ncel <= 50000:
        taxa.append(0.8)
    elif ncel > 50000 and ncel <= 100000:
        taxa.append(0.7)
    elif ncel > 100000 and ncel <= 200000:
        taxa.append(0.6)
    elif ncel > 200000 and ncel <= 500000:
        taxa.append(0.5)
    elif ncel > 500000:
        taxa.append(0.35)
```

```python
print(taxa)
```
